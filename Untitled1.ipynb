```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some dummy data (linear regression example)
np.random.seed(42)
X = np.random.rand(100, 1)  # 100 data points, single feature
y = 3 * X + 2 + np.random.randn(100, 1) * 0.1  # y = 3x + 2 + noise

# Add a bias term to X
X_b = np.c_[np.ones((X.shape[0], 1)), X]  # Add a column of ones for the bias term

# Initialize parameters
theta = np.random.randn(2, 1)  # 2 parameters: one for bias and one for the feature
learning_rate = 0.1
iterations = 1000
m = len(y)  # Number of data points

# Store loss for plotting
loss_history = []

# Batch Gradient Descent
for i in range(iterations):
    gradients = (2 / m) * X_b.T.dot(X_b.dot(theta) - y)  # Compute gradients
    theta -= learning_rate * gradients  # Update parameters
    
    # Compute and store loss
    loss = np.mean((X_b.dot(theta) - y) ** 2)
    loss_history.append(loss)

    # Optional: print progress every 100 iterations
    if i % 100 == 0:
        print(f"Iteration {i}: Loss = {loss:.4f}")

# Plot loss over iterations
plt.figure(figsize=(8, 5))
plt.plot(range(iterations), loss_history, label="Loss")
plt.xlabel("Iterations")
plt.ylabel("Loss (MSE)")
plt.title("Loss Over Iterations (Batch Gradient Descent)")
plt.legend()
plt.grid(True)
plt.show()

# Final model parameters
print("\nFinal parameters:")
print(f"Theta (bias): {theta[0][0]:.4f}")
print(f"Theta (slope): {theta[1][0]:.4f}")
```
