```python
class GDRegressor:

    def __init__(self,learning_rate,epochs):
        self.m= 77.96
        self.b= -120  #just random
        self.lr=learning_rate
        self.epochs=epochs

    def fit(self,x,y):

        for i in range(self.epochs):
            loss_slope= -2* np.sum(y- self.m*x.ravel() - self.b)
            self.b= self.b - (loss_slope*self.lr)
            print(self.m,self.b)

        print(self.b)
            
```
