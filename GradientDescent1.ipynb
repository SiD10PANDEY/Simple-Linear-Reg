```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
x,y = make_regression(n_samples=4, n_features=1, n_informative=1, noise=80,random_state=15)
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
#OLS
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='yellow')
```

```python
# Lets apply Gradient Descent assuming slope is constant m = 141.97
# and let's assume the starting value for intercept b = 0
y_pred = ((141.97 * x) + 0).reshape(4)          
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='yellow',label='OLS')
plt.plot(x,y_pred,color='#00a65a',label='b = 0')
plt.legend()
plt.show()
```

```python
m= 141.97
b= 0

loss_slope= -2* np.sum(y-m*x.ravel() - b)
loss_slope
```

```python
lr = 0.1  #assume

step_size= loss_slope * lr
step_size
```

```python
b=0
```

```python
b-= step_size
b
```

```python
y_pred1 = ((141.97 * x) + b).reshape(4)

plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='yellow',label='OLS')
plt.plot(x,y_pred1,color='pink',label='b= {}'.format(b))
plt.plot(x,y_pred,color='#00a65a',label='b = 0')
plt.legend()
plt.show()
```

```python
# Iteration 2
loss_slope = -2 * np.sum(y - m*x.ravel() - b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python

y_pred2 = ((141.97 * x) + b).reshape(4)

plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='OLS')
plt.plot(x,y_pred2,color='yellow',label='b = {}'.format(b))
plt.plot(x,y_pred1,color='#A3E4D7',label='b = {}'.format(b))
plt.plot(x,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

```python
# Iteration 3
loss_slope = -2 * np.sum(y - m*x.ravel() - b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python

y_pred3 = ((141.97 * x) + b).reshape(4)

plt.figure(figsize=(15,15))
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='OLS')
plt.plot(x,y_pred3,color='yellow',label='b = {}'.format(b))
plt.plot(x,y_pred2,color='#A3E4D7',label='b = {}'.format(b))
plt.plot(x,y_pred1,color='#A3E4D7',label='b = {}'.format(b))
plt.plot(x,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

```python
b=-100
m=141.97
lr=0.1
epochs=10

for epoch in range(epochs):
    loss_slope= -2* np.sum(y - m*x.ravel() - b)
    b= b - (loss_slope*lr)

    y_pred= m*x+b
    plt.plot(x,y_pred)

plt.plot(x,reg.predict(x),color='yellow', linestyle='dashed')
plt.scatter(x,y)
```
